```python
#Delete rows with missing values in all row
df.dropna(how = "all")
```

```python
#Delete rows with missing values in any part of the row
df.dropna(how="any")
```

```python
#Impute missing numerical values. Mean manipulation for specified column ex. "Fare" (index 152)

#first show rows with missing values
df[df["Fare"].isnull()]

#second replace nan value with mean
df.at[152, "Fare"] = df["Fare"].mean()

#check new value
df.iloc[152]
```

```python


#Impute missing numerical values. Median manipulation for specified column ex. "Fare" (index 152)

#first show rows with emtpy values
df[df["Fare"].isnull()]

#second, replace empty value for column Fare with the median of all vallues
df.at[152, "Fare"] = df["Fare"].median()

#check changes
df.iloc[152]
```

```python


#Impute missing numerical values. Mode manipulation for specified column ex. "Fare" (index 152)

#first, identify rows with empty values in "Fare" column
df[df["Fare"].isnull()]

#replace NaN with mode
df.at[152, "Fare"] = df["Fare"].mode().iloc[0]


#check results
df.iloc[152]
```

```python

#Drop columns with more than 15% percentage of missing values for specific column ex. "Cabin"
if df["Cabin"].isna().sum() / df["Cabin"].isna().count() > 0.15:
    df.drop("Cabin", axis= 1, inplace=True)

df.columns
```

```python


#Drop columns with more than 15% percentage of missing values for all columns
for column in df.columns:
    if df[column].isnull().sum() > 0.15:
        df.drop(labels=column, axis=1, inplace=True)
        

        
```

```python
#Handling Duplicates
#Detect and remove duplicate rows. Consider a group of columns.	Drop all except first occurrence

df.drop_duplicates(subset=["Age", "Cabin"], keep="first", inplace=True)

```

```python


#Detect and remove duplicate rows. Consider all columns.Drop all rows
df.drop_duplicates(inplace=True)

```

```python

#Detect and remove duplicate rows. Consider all columns.Drop all except last occurrence
df.drop_duplicates(inplace=True, keep="last")
```

```python

#Convert data types.Convert numeric columns to string
for column in df.columns:
    if df[column].dtype == "int64" or df[column].dtype == "float64":
        df[column] = df[column].astype(str)

df.info()
```

```python
import re

df = pd.read_csv("./datasets/titanic/titanic.csv")

def convert(value):
    value = re.findall("\d", value)
    value = "".join(value)
    return value

# takes the "Ticket" column from the Titanic dataset and standardizes it 
# all characters are in the same case (either uppercase or lowercase) for consistency. 
# Remove any non-alphanumeric characters to clean the ticket numbers.

df["Ticket"] = df["Ticket"].apply(lambda x: x.lower())
df["Ticket"] = df["Ticket"].apply(convert)

df["Ticket"]


```
